```python
beer_list = []
# loops through the number of styles. I found 170 different styles in the BreweryDB API
for Id in range (1,2):
    url = 'http://api.brewerydb.com/v2/beers?styleId='+str(Id)+'&key='+api_key
    data = requests.get(url).json()
    
    # loops through the number of pages
    for page in range(1,data['numberOfPages']+1):
        url2 = 'http://api.brewerydb.com/v2/beers?styleId='+str(Id)+'&p='+str(page)+'&key='+api_key
        data2 = requests.get(url2).json()
        
        # pulls individual beers
        for item in data2['data']:
            beer_info = [item.get('id'), item.get('name'),item.get('description'), item.get('abv'), item.get('ibu'),\
                         item.get('isOrganic'), item.get('available', {}).get('name'),item.get('styleId'), \
                         item.get('style', {}).get('name'), item.get('style', {}).get('shortName'),\
                         item.get('style', {}).get('description'), item.get('foodPairings'),\
                         item.get('labels', {}).get('medium')]
            beer_list.append(beer_info)
            
column_names = ['id', 'name', 'description', 'abv', 'ibu', 'isOrganic', 'availability', 'styleId', 'styleName',\
               'styleShortName', 'styleDescription', 'foodPairings', 'label']

df = pd.DataFrame(beer_list, columns=column_names)
df
```
