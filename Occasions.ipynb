```python
import pandas as pd
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:75% !important; }</style>"))
```

```python
beer_df=pd.read_csv('beers_subset_added_features.csv')
```

```python
beer_df1=pd.read_csv('beers_subset_wDishIngreds.csv')
```

```python
beer_df2=beer_df1.merge(beer_df,how='left', left_on='id', right_on='id')
```

```python
beer_df3=beer_df2.drop(['Unnamed: 0_x','Unnamed: 0.1','styleId_x','Unnamed: 0_y','name_y','description_y','abv_y','ibu_y','isOrganic_y','availability_y','styleId_y','styleName_y','styleShortName_y','styleDescription_y','foodPairings_y','label_y'], axis=1)
```

```python
Spring_Beers= ["Dubbel","Tribel","Garde","Doppelbock", "Miabock"]
Summer_Beers=["Weisse","Blonde Ale",'Cream Ale','Fruit Beer','KÃ¶lsch',"Saison",'Weizenbier','Witbier']
Fall_Beers=['Amber Ale','Brown Ale','India Pale Ale','Oktoberfest','Marzenbier','Rauchbier','Schwarzbier']
Winter_Beers=['Barley Wine','Strong Ale','Eisbock','Imperial Stout','Old Ale','Scotch Ale','Wassail']

```

```python
def fillSpringBeers():
    for j in Spring_Beers:
        beer_df3.loc[(beer_df3['styleName_x'].str.contains(j)), 'Occasion'] = 'Spring'
```

```python
fillSpringBeers()
```

```python
def fillSummerBeers():
    for j in Summer_Beers:
        beer_df3.loc[(beer_df3['styleName_x'].str.contains(j)), 'Occasion'] = 'Summer'
```

```python
fillSummerBeers()
```

```python
def fillFallBeers():
    for j in Fall_Beers:
        beer_df3.loc[(beer_df3['styleName_x'].str.contains(j)), 'Occasion'] = 'Fall'
```

```python
fillFallBeers()
```

```python
def fillWinterBeers():
    for j in Winter_Beers:
        beer_df3.loc[(beer_df3['styleName_x'].str.contains(j)), 'Occasion'] = 'Winter'
```

```python
fillWinterBeers()
```

```python
beer_df3.Occasion.replace(False, np.NaN, inplace=True)
```
