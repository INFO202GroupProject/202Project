```python
new_database = new_edited_df[['BeerId','Beer_Name','Beer_Description','ABV','IBU','srm','Is_Organic','Availability',\
                              'styleName','styleShortName','styleDescription','foodPairings','label',\
                              'Ingredients','Dishes','IngredientsDishes','fgMin','fgMax','ibuMin','ibuMax',\
                              'abvMin','abvMax','srmMin','srmMax','breweryId','breweryName',\
                              'brandClassification','country','locality','region','locationTypeDisplay','latitude',\
                              'longitude','Occassion','hex']]
new_database.columns = [['BeerId','Beer_Name','Beer_Description','ABV','IBU','SRM','Is_Organic','Availability',\
                              'styleName','styleShortName','styleDescription','foodPairings','label',\
                              'Ingredients','Dishes','IngredientsDishes','fgMin','fgMax','ibuMin','ibuMax',\
                              'abvMin','abvMax','srmMin','srmMax','breweryId','breweryName',\
                              'brandClassification','country','locality','region','locationTypeDisplay','latitude',\
                              'longitude','Occassion','hex']]
new_database
```

```python
copy_new_database = new_database
```

```python
abv_list = []
for abv in copy_new_database['ABV']:
    abv = round(float(abv),1)
    if abv <0.5:
        abv_list.append('alcohol-free')
    elif abv <1.5:
        abv_list.append('low-alcohol content')
    elif abv <4.0:
        abv_list.append('low/med-alcohol content')
    elif abv <7.5:
        abv_list.append('avg-alcohol content')
    elif abv <12.0:
        abv_list.append('high-alcohol content')
    elif abv >=12.0:
        abv_list.append('very high-alcohol content')
    else:
        abv_list.append('')
len(abv_list)
```

```python
copy_new_database['abvClass'] = abv_list
copy_new_database
```

```python
srm_dict = {2:'Pale Straw', 3: 'Straw', 4:'Pale Gold', 6:'Deep Gold', 7:'Light Amber', 9:'Pale Amber', 11:'Red Brown', \
       12:'Medium Amber', 14: 'Light Brown', 15:'Deep Amber', 17:'Chesnut',18:'Dark Red-Orange', 20: 'Brown',\
       24:'Ruby Brown', 30:'Deep Brown', 40:'Black' }
```

```python
import math
color_list = []
for srm in copy_new_database['SRM']:
    if math.isnan(srm):
        color_list.append('')
    else:
        srm = int(srm)
        if srm >= 1 and srm <=2:
            color_list.append('Pale Straw')
        elif srm <= 3:
            color_list.append('Straw')
        elif srm <=5:
            color_list.append('Pale Gold')
        elif srm <= 6:
            color_list.append('Deep Gold')
        elif srm <= 8:
            color_list.append('Light Amber')
        elif srm <= 10:
            color_list.append('Pale Amber')
        elif srm <= 11:
            color_list.append('Red Brown')
        elif srm <= 13:
            color_list.append('Medium Amber')
        elif srm <= 14:
            color_list.append('Light Brown')
        elif srm <= 16:
            color_list.append('Deep Amber')
        elif srm <= 17:
            color_list.append('Chesnut')
        elif srm <= 19:
            color_list.append('Dark Red-Orange')
        elif srm <= 22:
            color_list.append('Brown')
        elif srm <= 27:
            color_list.append('Ruby Brown')
        elif srm <= 36:
            color_list.append('Deep Brown')
        elif srm > 36:
            color_list.append('Black')
        else:
            color_list.append('')
len(color_list)
```

```python
copy_new_database['beerColor'] = color_list
```
