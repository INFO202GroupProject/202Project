#### Use these to build a map from beer style name to individual dishes and food categories. There are too many individual ingredients to list every time, so we use food categories. The ingredient lists are examples, and not exhaustive anyway. 

##### beer style ---> food categories & example dishes

```python
beer_ing_dishes = {}

def updateObjArr(obj, key, val):
    if key in obj:
        obj[key] = obj[key] + val
    else:
        obj[key] = val

def beer_to_ingr_dishes(beer_bflavors, bflavor_foodCats, foodCat_ings_dishes):
    
    for beer in beer_bflavors:
        if beer in beer_ing_dishes: 
            continue
        else:
            beer_ing_dishes[beer] = {}
        flavorList = beer_bflavors[beer]
        for flavor in flavorList:
            foodCategoryList = bflavor_foodCats[flavor]
            for foodCat in foodCategoryList:
                #ings = foodCat_ings_dishes[foodCat]['ingredients']
                updateObjArr(beer_ing_dishes[beer], 'ingredients', [foodCat])
            for foodCat in foodCategoryList:
                ds = foodCat_ings_dishes[foodCat]['dishes']
                updateObjArr(beer_ing_dishes[beer], 'dishes', ds)
```

## Merging Chart 1's Data with Existing Beer Data

##### Create new column in existing DataFrame (either one or two to keep Ingredients and Dishes separated). I went with combinin these so that we can just insert that into the foodPair columns that are missing data. 

```python
## Add Two new Columns 'Ingredients', 'Dishes'
def addFoodCols_sep():
    for beerStyle in beer_ing_dishes:
        ingredientList = beer_ing_dishes[beerStyle]['ingredients']
        ingStr = ' '.join(str(e) for e in ingredientList)
        dishList = beer_ing_dishes[beerStyle]['dishes']
        dishStr = ' '.join(str(f) for f in dishList)
        bl.loc[(bl['styleName'] == beerStyle), 'Ingredients'] = ingStr
        bl.loc[(bl['styleName'] == beerStyle), 'Dishes'] = dishStr

## Add One Column which contains 'Ingredients' + 'Dishes'
def addFoodCols():
    for beerStyle in beer_ing_dishes:
        ingredientList = beer_ing_dishes[beerStyle]['ingredients']
        ingStr = ' '.join(str(e) for e in ingredientList)
        dishList = beer_ing_dishes[beerStyle]['dishes']
        dishStr = ' '.join(str(f) for f in dishList)
        bl.loc[(bl['styleName'] == beerStyle), 'IngredientsDishes'] = ingStr + " " + dishStr

```
